```python
#判断50m
from math import radians, cos, sin, asin, sqrt

#公式计算两点间距离（m） # 经纬度转换成弧度
#lng1,lat1,lng2,lat2 = (120.12802999999997,30.28708,115.86572000000001,28.7427)
def geodistance(lng1,lat1,lng2,lat2):
    tmp_list=[float(lng1), float(lat1), float(lng2), float(lat2)]
    lng1, lat1, lng2, lat2 = map(radians, tmp_list)
    dlon=lng2-lng1
    dlat=lat2-lat1
    a=sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    distance=2*asin(sqrt(a))*6371*1000 # 地球平均半径，6371km
    distance=round(distance,3)
    if distance>50:
        return 0
    else:
        return 1 
```

```python
result['code'] = result.apply(lambda x: geodistance(x['上车经度'],x['上车纬度'],x['下车经度'],x['下车纬度']), axis=1)
#DEP_LONGITUDE	DEP_LATITUDE   DEST_LONGITUDE	DEST_LATITUDE
result.drop(index=(result.loc[result['code']==1].index),inplace=True)
result.shape
```

```python
import requests,json,time
def get_area(lng,lat):
    url='https://restapi.amap.com/v3/geocode/regeo?output=json&location={lng},{lat}&key=yourkey&radius=1000&extensions=all'
    res=requests.get(url.format(lng=lng,lat=lat)).text
    temp = json.loads(res)
    area= temp['regeocode']['formatted_address']
    time.sleep(0.5)
    print(area)
    return area
```
